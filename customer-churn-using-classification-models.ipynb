```python
import pandas as pd
import matplotlib.pyplot as plt

file_path = "/kaggle/input/telco-customer-churn/WA_Fn-UseC_-Telco-Customer-Churn.csv"
df = pd.read_csv(file_path)

churn_mapping = {"No": "No Churn", "Yes": "Churn"}
payment_mapping = {
    "Electronic check": "Electronic Check",
    "Mailed check": "Mailed Check",
    "Bank transfer (automatic)": "Bank Transfer (Automatic)",
    "Credit card (automatic)": "Credit Card (Automatic)"
}

inner_labels = ["No Churn", "Churn"]
inner_sizes = [df["Churn"].replace(churn_mapping).value_counts().loc[label] for label in inner_labels]
inner_colors = ["#a6cee3", "#fb9a99"]

payment_methods = ["Electronic Check", "Mailed Check", "Bank Transfer (Automatic)", "Credit Card (Automatic)"]
payment_data = df.groupby(df["Churn"].replace(churn_mapping)).apply(
    lambda x: x["PaymentMethod"].map(payment_mapping).value_counts()
).unstack(fill_value=0)

outer_sizes = []
outer_labels = []
for label in inner_labels:
    for pm in payment_methods:
        outer_sizes.append(payment_data.loc[label, pm])
        outer_labels.append(pm)

payment_color_mapping = {
    "Electronic Check": "#1f77b4",
    "Mailed Check": "#ff7f0e",
    "Bank Transfer (Automatic)": "#2ca02c",
    "Credit Card (Automatic)": "#d62728"
}
outer_colors = [payment_color_mapping[pm] for pm in outer_labels]

fig, ax = plt.subplots(figsize=(10, 7))
ax.axis('equal')
wedges1, _ = ax.pie(
    inner_sizes, radius=0.7, labels=inner_labels, colors=inner_colors,
    startangle=90, wedgeprops=dict(width=0.6, edgecolor='white'),
    textprops={'fontsize': 12, 'weight': 'bold'}
)
wedges2, _ = ax.pie(
    outer_sizes, radius=1.0, labels=[""] * len(outer_sizes), colors=outer_colors,
    startangle=90, wedgeprops=dict(width=0.3, edgecolor='white')
)
plt.legend(wedges2[:4], payment_methods, title="Payment Methods", bbox_to_anchor=(1, 0.5), loc="center left")
plt.title("Customer Payment Method with Respect to Churn", fontsize=16, color="darkblue", pad=20)
plt.tight_layout()
plt.show()
```

## 4. preprocessing

```python
df.replace('No internet service', 'No', inplace=True)
```

```python
df.drop('customerID',axis=1,inplace=True)
```

```python
label_encoders = {}

for col in df.select_dtypes(include=['object', 'category']).columns:
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col])
    label_encoders[col] = le
```

```python
df.head()
```

df.isna().sum()


```python
imputer = SimpleImputer(strategy='mean')
df['TotalCharges'] = imputer.fit_transform(df[['TotalCharges']])
```

```python
df.isna().sum()
```

```python
def remove_outliers(df, col):
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return df[(df[col] >= lower_bound) & (df[col] <= upper_bound)]

for col in ["tenure", "MonthlyCharges", "TotalCharges"]:
    df = remove_outliers(df, col)
    
```

```python
scaler = StandardScaler()
df[["tenure", "MonthlyCharges", "TotalCharges"]] = scaler.fit_transform(
    df[["tenure", "MonthlyCharges", "TotalCharges"]])

print(df[["tenure", "MonthlyCharges", "TotalCharges"]].head())
```

## 5. Modeling

```python
X = df.drop(['Churn'], axis=1)
y = df['Churn']

smote = SMOTE(random_state=42)
X_res, y_res = smote.fit_resample(X, y)
X_train, X_test, y_train, y_test = train_test_split(X_res, y_res, test_size=0.2, random_state=42)
```

#### 4.1 LogisticRegression

```python
lr_model = LogisticRegression(max_iter=1000)
lr_model.fit(X_train, y_train)

y_pred = lr_model.predict(X_test)

print("Accuracy:", accuracy_score(y_test, y_pred))
print("\n")
print("Classification Report:\n", classification_report(y_test, y_pred))
```

```python
cm = confusion_matrix(y_test, y_pred)

plt.figure(figsize=(15, 7))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', cbar=False)
plt.xlabel("Predicted Labels", fontsize=14)
plt.ylabel("Actual Labels", fontsize=14)
plt.title("Confusion Matrix", fontsize=16, pad=25)
plt.tight_layout()
plt.show()
```

##### 4.2 XGBClassifier

```python
xgb_model = XGBClassifier(random_state=42, use_label_encoder=False, eval_metric='logloss')

xgb_model.fit(X_train, y_train)

y_pred_xgb = xgb_model.predict(X_test)

print("XGBoost Accuracy:", accuracy_score(y_test, y_pred_xgb))
print("\n\n")
print("XGBoost Classification Report:\n", classification_report(y_test, y_pred_xgb))
```

```python
cm_xgb = confusion_matrix(y_test, y_pred_xgb)

plt.figure(figsize=(15, 7))
sns.heatmap(cm_xgb, annot=True, fmt='d', cmap='Blues', cbar=False)
plt.xlabel("Predicted Labels", fontsize=14)
plt.ylabel("Actual Labels", fontsize=14)
plt.title("XGBoost Confusion Matrix", fontsize=16, pad=20)
plt.tight_layout()
plt.show()
```

##### 4.3 Random ForestClassifier

```python
rf_model = RandomForestClassifier(random_state=42)
rf_model.fit(X_train, y_train)

y_pred_rf = rf_model.predict(X_test)

print("RandomForest Accuracy:", accuracy_score(y_test, y_pred_rf))
print("\nClassification Report:\n", classification_report(y_test, y_pred_rf))
```

```python
cm_rf = confusion_matrix(y_test, y_pred_rf)

plt.figure(figsize=(15, 7))
sns.heatmap(cm_rf, annot=True, fmt='d', cmap='Blues', cbar=False)
plt.xlabel("Predicted Labels", fontsize=14)
plt.ylabel("Actual Labels", fontsize=14)
plt.title("Random Forest Confusion Matrix", fontsize=16, pad=20)
plt.tight_layout()
plt.show()
```
